```python
import pandas as pd ## import the pandas library

board_exam = pd.read_excel('board2.xlsx') ##  to load the excel file into a Data Frame
board_exam
```

```python
Instru = board_exam.loc[(board_exam['Electronics'] >70) &  ## filtering the students who got a score higher than 70 in electronics
(board_exam['Track'] == 'Instrumentation') & ## filtering the students from Instrumentation Track
(board_exam['Hometown'] == 'Luzon'), ## filtering the students from Luzon
["Name", "GEAS", "Electronics"]] ## displaying 3 specific columns only

Instru
```

```python
## calculating the average score across 4 subjects
board_exam["Average"] = board_exam[["Math", "Electronics", "GEAS", "Communication"]].mean(axis=1) #to calculate the mean and stored in a row 
board_exam ## to display the new Data Frame with average row
```

```python
Mindy = board_exam.loc[(board_exam['Average'] >=55)& ## filtering the students who got 55 and up average
        (board_exam['Gender'] == 'Female')& ## filtering the female students
        (board_exam['Hometown'] == 'Mindanao'), ## fiktering to those whose hometown is Mindanao
         ["Name", "Track", "Electronics", "Average"]] ## To display only 4 columns

Mindy
```
