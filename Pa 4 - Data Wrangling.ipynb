```python
import numpy as np

X = np.random.random((5,5))

X_Normalized = (X - X.mean())/X.std()
np.save('X_normalized.npy', X_Normalized)
print(X_Normalized)
```

```python
import numpy as np

A = np.array([i**2 for i in range(1, 101)])
A.reshape((10,10))

divisible_by_3 = A[A%3 == 0]

np.save('div_by_3', divisible_by_3)
print(divisible_by_3)
```
